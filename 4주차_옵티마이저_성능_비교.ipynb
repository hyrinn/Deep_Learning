```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier, KerasRegressor
import os
import numpy as np
```

```python
# мӢӨн—ҳ кІ°кіј мһ¬нҳ„мқ„ мң„н•ң seed кі м •
SEED = 4
os.environ['PYTHONHASHSEED'] = str(SEED)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
tf.random.set_seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)
initializer = tf.keras.initializers.GlorotUniform(seed=SEED)
```

```python
# лҚ°мқҙн„°м„ёнҠё мӨҖл№„
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
```

```python
# лҚ°мқҙн„° м „мІҳлҰ¬
train_images = train_images.reshape((60000, 784))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 784))
test_images = test_images.astype('float32') / 255

train_labels = tf.keras.utils.to_categorical(train_labels)
test_labels = tf.keras.utils.to_categorical(test_labels)
```

```python
# мӢ кІҪл§қ лӘЁлҚё кө¬м¶•
def build_model():
    network = tf.keras.models.Sequential()

    # мқҖлӢүмёө мғқм„ұ
    # м—¬кё°м—җ мҪ”л“ңлҘј мһ‘м„ұн•ҙмЈјм„ёмҡ”

    network.add(tf.keras.layers.Dense(512, activation='relu', input_shape=(784,),kernel_initializer=initializer))
    network.add(tf.keras.layers.Dense(10, activation='sigmoid', kernel_initializer=initializer))

    # мҳөнӢ°л§Ҳмқҙм Җ м„Өм • н•„мҡ”
    network.compile(optimizer='adam',
                   loss='mse',
                   metrics=['accuracy'])
    return network
```

```python
# model мғқм„ұ
model = build_model()
```

```python
result = model.fit(train_images, train_labels, epochs=5, batch_size=64)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
```

```python
print('н…ҢмҠӨнҠё м •нҷ•лҸ„ : ', test_acc)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adagrad, Adadelta, RMSprop, Adam
import pandas as pd
import numpy as np
import os
```

```python
# лҚ°мқҙн„° load
dataset = pd.read_csv(os.path.join(data_path, '/content/drive/MyDrive/diabetes.csv'))

X = dataset.iloc[:, 0:8].values
y = dataset.iloc[:, 8].values
```

```python
# мҲҳм • x
def configure_tf_determinism():
    # TensorFlowмқҳ лі‘л ¬м„ұ к°җмҶҢ
    tf.config.experimental.enable_op_determinism()
```

```python
# мӢӨн—ҳ кІ°кіј мһ¬нҳ„мқ„ мң„н•ң seed кі м •
SEED = 4
os.environ['PYTHONHASHSEED'] = str(SEED)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
tf.random.set_seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)
initializer = tf.keras.initializers.GlorotUniform(seed=SEED)
```

```python
# лӘЁлҚё мғқм„ұ
def build_model():
    model = Sequential()
    # мҲҳм • к°ҖлҠҘ
    model.add(tf.keras.layers.Dense(12, input_dim=8, activation='sigmoid', kernel_initializer=initializer))
    model.add(tf.keras.layers.Dense(8, activation='sigmoid', kernel_initializer=initializer))

    # мҲҳм • x
    model.add(tf.keras.layers.Dense(1, activation='sigmoid', kernel_initializer=initializer))

    model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

    return model
```

```python
# мҲҳм • x
configure_tf_determinism()
```

```python
model = build_model()
```

```python
model.fit(X, y, epochs=30, batch_size=40)
```

```python
_, accuracy = model.evaluate(X, y)
```

```python
print('м •нҷ•лҸ„: %.2f' % (accuracy*100))
```
